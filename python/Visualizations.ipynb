```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/df_labeled_filtered.csv")
```

```python
# Define the order of the labels
order = ['n', '0', 'p']

# Define the colors for each label
palette = {'n': 'red', '0': 'blue', 'p': 'green'}

# Plot the value counts with the specified order and colors
ax = sns.countplot(x='label', data=df, order=order, palette=palette)

# Add the values on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, -15), textcoords='offset points')

plt.xlabel('Label')
plt.ylabel('Count')
plt.title('Distribution of Labels before balancing')
plt.show()
```

```python
# limit the amount of values to 2066 for each label 
df = df.groupby('label').apply(lambda x: x.sample(2066)).reset_index(drop=True)
```

```python
# Define the order of the labels
order = ['n', '0', 'p']

# Define the colors for each label
palette = {'n': 'red', '0': 'blue', 'p': 'green'}

# Plot the value counts with the specified order and colors
ax = sns.countplot(x='label', data=df, order=order, palette=palette)

# Add the values on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, -15), textcoords='offset points')

plt.xlabel('Label')
plt.ylabel('Count')
plt.title('Distribution of Labels after balancing')
plt.show()
```
